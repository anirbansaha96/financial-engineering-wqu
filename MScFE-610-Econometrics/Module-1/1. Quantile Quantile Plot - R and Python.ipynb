# Quantile-Quantile Plot

#### A Quantile-Quantile Plot is a graphical technique to check if two datasets come from propulations with the same distribution.

### A Quantile-Quantile Plot is a plot of quantiles of one dataset against the quantiles of another dataset. A 45 degree line is also plotted. 

#### If the two datasets come from a population with the same distribution, the points fall approximately along this 45 degree reference line.

Note: 
1) Quantile := Fraction of datapoints below the given value. For example, 0.3 quantile is the point where 30% of the sorted data falls below the point and 70% of the sorted data is above the point. 

2) The greater the deviation of the points from the 45 degree reference line, the greater is the evidence against the hypothesis that they come from the same population.

3) If both the datasets have the same size, the Q-Q plot is essentially a plot of sorted values of one dataset against the sorted values of the other dataset. If the datasets are of unequal size, quantiles are picked from the sorted values of the smaller dataset and the corresponding quantiles from the larger dataset are plotted. 

## In a Normal Quantile-Quantile Plot, one of the datasets is the normal distribution. The Normal QQ Plot is an important statistical tool, as it is one of the easiest graphical methods to test for normality of a dataset, which is an essential assumption for many statistical tests. 

We will now show how to check if two datasets come from the same population using R programming Language:

```python
dataset1=rnorm(1000) #This generates a dataset with 1000 points from the standard normal distribution
```

```python
qqnorm(dataset1)     #This plots the Normal QQ Plot for the dataset1
qqline(dataset1)
```

Here we see that our dataset1 follows the normal distribution since the quantiles approximately lie on the reference line.

```python
dataset2=rcauchy(500, location=0, scale=1)    #This generates a dataset2 with 500 points from the Cauchy Distribution
```

```python
qqnorm(dataset2)
qqline(dataset2)
```

Comparing the plot we can infer that it doesn't follow the Normal Distribution.

```python
qqplot(x=dataset1, y=dataset2)
```

Again we can see that dataset1 and dataset2 do not come from populations with the same distribution.

Now we do the same thing in Python Programming Language:

```python
import numpy as np
dataset1=np.random.normal(0,1,1000) #Generates a dataset1 with 1000 points drawn  from a Normal(0,1) distribution
dataset2 = np.random.standard_cauchy(500) #Generates dataset2 with 1000 points drawn  from a Cauchy distribution with mode=0
```

```python
import statsmodels.api as sm
import pylab as py
```

```python
sm.qqplot(dataset1,line='45')
py.show()
```

```python
sm.qqplot(dataset2,line='45')
py.show()
```

Hence we can see that dataset1 follows a normal distribution byt dataset2 does not.

Note: 
1) A steep slope for a QQ plot (greater than 45 degrees) imply, that the observations are more spread out than if they were from a normal distribution.
2) A flatter slope for a QQ plot (less than 45 degrees) imply, that the observations are less spread out than if they were from a normal distribution.
